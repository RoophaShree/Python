Generating a file containing Email,phone number,Home address, name,year born,no.of kids,home status,income,tickets

https://www.ssa.gov/oact/babynames/   --  Source from which namelist was created

```python
import random
import pandas as pd
import os
from random import randint
names=open("namelist.txt","r")
names1=names.read()
```

```python
name=names1.split()
email_list=["gamil.com","outlook.com"]
house=["own", "rent"]
```

https://state.1keydata.com/ -- Source from which usa_states was created

```python
def house_address():
    zip_code=str(randint(10000,99999))
    states=open("usa_states.txt","r")
    stateslist=states.read()
    statelist=stateslist.split()
    flatno=str(randint(200,500))
    house_address="flatno"+" "+str(flatno)+" "+str(random.choice(statelist))+" "+"USA"+zip_code
    return house_address
```

choice() is a inbuilt function in Python that returns a random item from a list, tuple or string

randint() is used to generate random numbers which is randint(start,end)

append() is used to add an item to the end of the list

```python
com_list=[]
for x in range (100):
    name1=random.choice(name)
    email=str(name1.replace(" "," "))+"@"+random.choice(email_list)
    mobileno=random.randint(1000000000,9999999999)
    year=str(randint(1960,2019))
    kids=str(randint(1,4))
    home=random.choice(house)
    annual=random.randint(999,999999)
    tickets=random.randint(1,30)
    my_list=[]
    my_list.append(email)
    my_list.append(mobileno)
    my_list.append(house_address())
    my_list.append(name1)
    my_list.append(year)
    my_list.append(kids)
    my_list.append(home)
    my_list.append(annual)
    my_list.append(tickets)
    com_list.append(my_list)
com_list
```

List is loaded to dataframe and columns are defined directly from the list

```python
df=pd.DataFrame(com_list)
df.columns=['Email address','Phone Number','Home Address','Persons Name','Year Born','No. of Kids','Home Status','Annual','Number of Speeding Tickets']
df
```

The below is used to convert the dataframe to csv and stores it in our system

```python
df.to_csv('fakeName.csv')
```
