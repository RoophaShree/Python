Imagine that you are starting a new Pizza restaurant franchise and you would like to able to take online orders.

Create a python class Pizza Order that has following capabilities:

- Ability to store and retrieve customer first name and last name
- Ability to store and retrieve the order date
- Ability to store and retrieve the order amount
- Ability to arrange the order by order amount.


Creating class PizzaOrder which stores customer firstname, Lastname, order date and order amount

```python
import datetime
class PizzaOrder(object):
    def __init__(self,name):
        self.name = name
        first = name.index(' ')
        self.first_name = name[:first]
        try:
            last=name.rindex(' ')
            self.last_name=name[last+1:]
        except:
            self.last_name=name
            
    def get_name(self):
        return self.name
    
    def get_lastname(self):
        return self.last_name 
    
    def get_firstname(self):
        return self.first_name
    
    def set_orderdate(self,orderdate):
        self.order_date=orderdate
        
    def get_orderdate(self):
        return self.order_date
    
    def set_orderamount(self, orderamount):
        self.orderamount=orderamount
        
    def get_orderamount(self):
        return self.orderamount
        
    def __lt__(self,other):
        return self.orderamount<other.orderamount
    
    def __str__(self):
        return self.name
    
    
    
```
